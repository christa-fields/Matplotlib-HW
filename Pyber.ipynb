```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load CSV
city_data ="Pyber/raw_data/city_data.csv"
ride_data ="Pyber/raw_data/ride_data.csv"
```

```python
# Read ride_data with pandas
ride_data_df = pd.read_csv(ride_data)
ride_data_df.head()
```

```python
# Read city_dta with pandas
city_data_df = pd.read_csv(city_data)
city_data_df.head()
```

```python
#merge the two tables
merge_table = pd.merge(ride_data_df, city_data_df, on= "city")
merge_table.head()
```

```python
#Urban
urban = merge_table.loc[merge_table["type"] =="Urban"]
urban_table_df=urban.groupby(["city"])
urban_table_df.max()
```

```python
#Suburban
suburban = merge_table.loc[merge_table["type"] =="Suburban"]
suburban_table_df= suburban.groupby(["city"])
suburban_table_df.max()
```

```python
#Rural
rural = merge_table.loc[merge_table["type"] =="Rural"]
rural_table_df= rural.groupby(["city"])
rural_table_df.max()
```

```python
#Urban average fare
urban_average_fare=pd.DataFrame(urban_table_df["fare"].mean())
urban_average_fare.reset_index(inplace=True)
urban_average_fare.columns=["city", "urban_average_fare"]
urban_average_fare.head()
```

```python
#Suburban average fare
suburban_average_fare=pd.DataFrame(suburban_table_df["fare"].mean())
suburban_average_fare.reset_index(inplace=True)
suburban_average_fare.columns=["city", "suburban_average_fare"]
suburban_average_fare.head()
```

```python
#Rural average fare
rural_average_fare=pd.DataFrame(rural_table_df["fare"].mean())
rural_average_fare.reset_index(inplace=True)
rural_average_fare.columns=["city", "rural_average_fare"]
rural_average_fare.head()
```

```python
#Urban total rides per city

urban_total_rides = pd.DataFrame(urban_table_df["ride_id"].count())
urban_total_rides.reset_index(inplace=True)
urban_total_rides.columns=["city", "urban_total_rides"]
urban_total_rides.head()
```

```python
#Suburban total rides per city

suburban_total_rides = pd.DataFrame(suburban_table_df["ride_id"].count())
suburban_total_rides.reset_index(inplace=True)
suburban_total_rides.columns=["city", "suburban_total_rides"]
suburban_total_rides.head()
```

```python
#Rural total rides per city

rural_total_rides = pd.DataFrame(rural_table_df["ride_id"].count())
rural_total_rides.reset_index(inplace=True)
rural_total_rides.columns=["city", "rural_total_rides"]
rural_total_rides.head()
```

```python
#Urban total driver per city
urban_drivers=pd.DataFrame(urban_table_df["driver_count"].sum())
urban_drivers.reset_index(inplace=True)
urban_drivers.columns=["city", "urban_total_drivers"]
urban_drivers.head()
```

```python
#Suburban total driver per city
suburban_drivers=pd.DataFrame(suburban_table_df["driver_count"].sum())
suburban_drivers.reset_index(inplace=True)
suburban_drivers.columns=["city", "suburban_total_drivers"]
suburban_drivers.head()
```

```python
#Rural total driver per city
rural_drivers=pd.DataFrame(rural_table_df["driver_count"].sum())
rural_drivers.reset_index(inplace=True)
rural_drivers.columns=["city", "rural_total_drivers"]
rural_drivers.head()
```

```python
# Merge the urban created data frames on the name of the city
urban_merged_1 = pd.merge(urban_average_fare, urban_total_rides, on="city")
urban_merged_df=pd.merge(urban_merged_1, urban_drivers, on="city")
urban_merged_df.head()
```

```python
# Merge the suburban created data frames on the name of the city
suburban_merged_1 = pd.merge(suburban_average_fare, suburban_total_rides, on="city")
suburban_merged_df=pd.merge(suburban_merged_1, suburban_drivers, on="city")
suburban_merged_df.head()
```

```python
# Merge the rural created data frames on the name of the city
rural_merged_1 = pd.merge(rural_average_fare, rural_total_rides, on="city")
rural_merged_df=pd.merge(rural_merged_1, rural_drivers, on="city")
rural_merged_df.head()
```

```python
c_urban='gold'
c_suburban='lightskyblue'
c_rural='lightcoral'

size_urban=urban_merged_df["urban_total_drivers"]
size_suburban=suburban_merged_df["suburban_total_drivers"]
size_rural=rural_merged_df["rural_total_drivers"]

urban_merged_df.plot(kind="scatter",x="urban_total_rides", y="urban_average_fare", s=size_urban, marker='o', c=c_urban)
suburban_merged_df.plot(kind="scatter", x="suburban_total_rides", y="suburban_average_fare", s=size_suburban, marker='o', c=c_suburban)
rural_merged_df.plot(kind="scatter", x="rural_total_rides", y="rural_average_fare", s=size_rural, marker='o', c=c_rural)

    
# Create a title, x label, and y label for our chart
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average Fare ($)")

plt.show()
```
